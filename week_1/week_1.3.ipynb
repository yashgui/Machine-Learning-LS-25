```python
import pandas as pd
import numpy as np
```

```python
names = ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah', 'Ian', 'Jane']
subjects = ['Math', 'Science', 'English', 'Math', 'Science', 'English', 'Math', 'Science', 'English', 'Math']
scores = np.random.randint(50, 101, size=10)

df = pd.DataFrame({'Name': names,'Subject': subjects,'Score': scores,'Grade': ''})
df 
```

```python
def assign_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

df['Grade'] = df['Score'].apply(assign_grade)
df 
```

```python
#Sorted by Score (descending)
df.sort_values(by='Score', ascending=False)
```

```python
df1= df.groupby(['Subject']).agg({'Score':'mean'})
df1 
```

```python
def pandas_filter_pass(df):
    return df[df['Grade'].isin(['A','B'])].sort_values('Grade')
pandas_filter_pass(df)
```
